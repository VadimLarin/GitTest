```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
class Chatbot:
    def __init__(self, model_name):
        self.model = GPT2LMHeadModel.from_pretrained(model_name)
        self.tokenizer = GPT2Tokenizer.from_pretrained(model_name)

    def generate_response(self, user_input, max_length=100):
        input_ids = self.tokenizer.encode(user_input, return_tensors="pt")
        outputs = self.model.generate(input_ids, max_length=max_length, num_beams=5, no_repeat_ngram_size=2, top_k=50, top_p=0.95, temperature=0.7)
        response = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
        return response
```

```python
chatbot = Chatbot("sberbank-ai/rugpt3large_based_on_gpt2")
```

```python
user_question = "Какой смысл жизни?"
response = chatbot.generate_response(user_question)
print(f"Ответ: {response}")
```
