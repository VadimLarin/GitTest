```python
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
import spacy
spacy.prefer_gpu()
nlp = spacy.load('ru_core_news_sm')
import gensim.downloader
from gensim.models import KeyedVectors
word2vec_rus = gensim.downloader.load('word2vec-ruscorpora-300')
```

```python
class PreprocessorText:
      def load_file(self):
        with open("book1863_ok.txt", "r", encoding="utf-8") as file:
            self.text1 = file.read()

      def tokenized_words(self):
        self.tokenized = nltk.word_tokenize(self.text1)

      def filter_words(self):
        words = self.nlp(self.text1)
        self.word_filt = [token.text for token in words if not token.is_punct and not token.is_stop]

      def lemmatized_words(self):
        words = self.nlp(self.text1)
        self.word_learn = [token.lemma_ for token in words if not token.is_punct and not token.is_stop]

      def __init__(self):
        self.nlp = spacy.load("ru_core_news_sm")
        self.word2vec_rus = gensim.downloader.load('word2vec-ruscorpora-300')

      def word2vec_text(self, num_word):
        for index, word in enumerate(self.word2vec_rus.index_to_key):
            if index == num_word:
                break
            print(f"word#{index}/{len(self.word2vec_rus.index_to_key)} is {word}")

processor_text1 = PreprocessorText()
processor_text1.load_file()
processor_text1.tokenized_words()
processor_text1.filter_words()
processor_text1.lemmatized_words()
processor_text1.word2vec_text(100)
```
